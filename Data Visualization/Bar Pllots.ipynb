```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from PIL.FontFile import WIDTH
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Sachin Tendulkar Runs')
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
x = np.arange(len(years))
width = 0.25
plt.bar(x-width, kohli, WIDTH, label='Kohli')
plt.bar(x+width, sehwag, width, label='Sehwag')
plt.bar(x, sachin, width, label='Sachin')

plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Runs Comparison')
plt.xticks(x, years)
plt.legend()
plt.show()
```

Horizontal Bar Chart

```python
players = ["Sachin", "Sehwag","Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+1100+900]
plt.barh(players, runs_5yrs, color='pink')
plt.xlabel('Total Runs in first 5 years')
plt.title("First 5 years runs comparison")
plt.tight_layout()
plt.show()
```

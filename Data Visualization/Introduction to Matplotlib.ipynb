```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years, kohli, color="green", linestyle="--", linewidth=3 , label='Kohli')
plt.plot(years, sehwag, color= "yellow", linestyle="--", linewidth=3 , label='Sehwag')
plt.plot(years, sachin, color="red", linestyle="--", linewidth=3 , label='Sachin')

plt.xlabel('Years')
plt.ylabel('Runs')
plt.title("Kholi v/s Sehwag v/s Sachin")
plt.grid(True)
plt.legend()
plt.show()
```
